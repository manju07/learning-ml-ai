## Random 

Numpy also has lots of ways to create random number arrays:

### rand
Create an array of the given shape and populate it with
random samples from a uniform distribution
over ``[0, 1)``.

````python

# A uniform distribution is a type of probability distribution in which all values within a certain range are equally likely to occur.
# In NumPy, np.random.rand generates numbers from a uniform distribution over [0, 1), meaning every number in this interval has an equal chance of being selected.

# np.random.rand(2) generates a 1-dimensional array of 2 random numbers sampled from a uniform distribution over [0, 1)
np.random.rand(2)
````

````python
# np.random.rand(5,5) creates a 5x5 array of random numbers drawn from a uniform distribution over [0, 1)
np.random.rand(5,5)
````

### randn

Return a sample (or samples) from the "standard normal" distribution. Unlike rand which is uniform:

#### The standard normal distribution is a special kind of normal (Gaussian) distribution with:
- **Mean of 0** (center of the distribution, i.e., the peak of the bell curve)
- **Standard deviation of 1** (controls the width/dispersion of the curve)

When drawing samples from a standard normal distribution:
- Most values will be close to 0
- Some values are negative, some are positive
- Extreme values (far from 0) are rare

**Example usage:**  
Create an array of 8 random numbers drawn from a standard normal distribution:
```python
data = np.random.randn(8)
print(data)
# Example output:
# array([-1.12,  0.76, -0.22,  0.87,  0.14, -0.70,  1.48, -0.96])
```

Notice how the numbers are centered around zero, and both positives and negatives are possible, with most numbers not too far from 0.

````python

# The standard normal distribution is a special case of the normal (Gaussian) distribution with a mean of 0 and a standard deviation of 1.
# In a standard normal distribution:
#   - The mean = 0 (center of the "bell curve")
#   - The standard deviation = 1 (spread/width of the bell curve)
# Most values fall near 0, with fewer values farther from 0, following the familiar bell-shaped curve.
# Values can be positive or negative, and the probability of extreme values decreases rapidly as you move away from 0.

# np.random.randn(2) generates a 1-dimensional array with 2 random numbers drawn from a standard normal distribution (mean 0, variance 1).
# How it works: For each element, NumPy samples from a "bell curve" (Gaussian/normal distribution), so most values will be near 0, but some can be higher or lower.

np.random.randn(2)
````

````python
# np.random.randn(5, 5) generates a 5x5 array where each value is drawn from a standard normal distribution (mean 0, variance 1).
# That means each number is randomly chosen according to the bell curve of the normal distribution.
np.random.randn(5, 5)
````
